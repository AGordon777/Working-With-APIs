### World Food Info - Jan also did this, so I added a fourth at bottom

https://world.openfoodfacts.org/product/2000000002603/big-mac-mcdonald-s

Write in your barcode and find info about the food.

```python
barcode = '2000000002603'

response = requests.get('https://world.openfoodfacts.org/api/v0/product/'+barcode+'.json')
response
```

```python
results = response.json()

flattened_data = json_normalize(results).T

display(flattened_data.shape)
flattened_data.head(30)
```

### Dictionary - type in a word and see what it means

https://api.dictionaryapi.dev/api/v2/entries/en/

```python
word = 'panda'
response4 = requests.get('https://api.dictionaryapi.dev/api/v2/entries/en/'+word)
response4
```

```python
results4 = response4.json()
results4

flattened_data4 = json_normalize(results4)

word_types=[]
definitions=[]

for el in flattened_data4['meanings'][0]:
    for d in el['definitions']:
        definitions.append(d['definition'])
        word_types.append(el['partOfSpeech'])

words = pd.DataFrame({"Type":word_types,"Definition":definitions})
words['Word'] = word
words = words[['Word','Type','Definition']]
words
```
